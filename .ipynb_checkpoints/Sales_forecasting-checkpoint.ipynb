## Notebook for Kaggle competition: Forecasting sales

### - Import libraries -

```python
import numpy as np     
import pandas as pd       
import seaborn as sns   
import matplotlib.pyplot as plt         
import sys

from scipy import stats
from sklearn.metrics import mean_squared_log_error, mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from statsmodels.tsa.deterministic import CalendarFourier, DeterministicProcess
from scipy.signal import periodogram
from xgboost import XGBRegressor
from myFunctions import myPrint
```

### - Import the data - 

#### Read in train dataset

```python
df_train = pd.read_csv('train.csv', parse_dates=['date'], infer_datetime_format=True)
df_train['date'] = df_train.date.dt.to_period('D')

print(df_train.columns.values)
```

#### Analyse the stores and family columns

```python
product_types=df_train['family'].unique()

print(product_types)
print(len(product_types))

number_of_stores=df_train['store_nbr'].unique()
print(np.sort(number_of_stores))
```

#### Read in test dataset

```python
df_test = pd.read_csv('test.csv', parse_dates=['date'])
df_test['date'] = df_test.date.dt.to_period('D')
```

#### Read in holiday data set and check for missing values

```python
holiday = pd.read_csv('holidays_events.csv', parse_dates=['date'])
holiday = holiday.set_index('date').to_period('D')
myPrint(holiday.isna().sum(), printmessage="Missing Holidays", hold=False)
```

#### Read in oil data set and check for missing values

```python
oil = pd.read_csv('oil.csv', parse_dates=['date'])
oil = oil.set_index('date').to_period('d')

myPrint(oil.head(10), printmessage="oils", hold=False)
myPrint(oil.isna().sum(), printmessage="Missing Oil----", hold=False)
```

#### Resample is because there are missing dates in the data

```python
# Back fill the missing values
oils = oil.resample('d').mean().fillna(method='bfill') 
myPrint(oils.head(10), printmessage="Newoils3", hold=False)

myPrint(oils.isna().sum(), printmessage="Missing Oils", hold=False)

## Create a weekly average. Note first 6 values will be NaN due to window size
myPrint(oils.rolling(7).mean().head(10), hold=False)
```

#### Create full data set

```python
full_df = pd.concat([df_train, df_test])
full_df.reset_index(drop=True, inplace=True)
```

#### Create store data containing the store number, family and date labels

```python
X_store = full_df.set_index(['store_nbr', 'family', 'date']).sort_index()
myPrint(X_store.head(), printmessage="Xstore head Data", hold=False)
```

#### Extract sales labels

```python
y_ = X_store.loc[:,:,'2017':].dropna()['sales'].to_frame().unstack(['store_nbr', 'family'])
myPrint(y_, printmessage="labels", hold=False)
```

#### Calculate average sales

```python
av_sales = (X_store.groupby('date').mean().squeeze())['sales'].to_frame().dropna()
myPrint(av_sales, printmessage="averagesales", hold=False)
```

#### Plot periodogram of average sales as rough indicator of what Fourier components to use

```python
fs = pd.Timedelta("1Y") / pd.Timedelta("1D")
myPrint(fs, hold=False)
frequencies, spectrum = periodogram(av_sales.sales, fs=fs, detrend='linear', window="boxcar", scaling='spectrum')

_, ax = plt.subplots()
ax.step(frequencies, spectrum, color="purple")
ax.set_xscale("log")
ax.set_xticks([1, 2, 4, 6, 12, 26, 52, 104])
ax.set_xticklabels(["Annual (1)","Semiannual (2)","Quarterly (4)","Bimonthly (6)",
                    "Monthly (12)","Biweekly (26)","Weekly (52)","Semiweekly (104)"],rotation=30,)
ax.ticklabel_format(axis="y", style="sci", scilimits=(0, 0))
ax.set_ylabel("Variance")
ax.set_title("Periodogram")
plt.show()
```

#### Create monthly Fourier components and dummies for days of the week

```python
fourier = CalendarFourier(freq='M', order=4) 
## Monthly seasonality. 4 frequency components chosen so Fourier captures weekly changes. 
## Use dummy variables for the seasonality (daily changes in a week - so dummies for each day)
## Could use Fourier for entire series, but would need 28 frequency components to model 

data_indices = X_store.loc[:,:,:].unstack(['store_nbr', 'family']).index
dp = DeterministicProcess(
    index=data_indices,
    constant=True,
    order=1, # Fit order
    seasonal=True,
    additional_terms=[fourier],
    drop=True,
)
X_full = dp.in_sample()

## Feature columns
myPrint(X_full.columns.values, printmessage="Feature columns", hold=False)
```

### - Create other features -

#### Creat a new years day feature

```python
X_nyd = X_full.copy()
X_nyd['new_year'] = (X_nyd.index.dayofyear ==1).astype('int') 
```

#### Create pay day feature

```python
X_pay = X_nyd.copy()

## Create a temporary days in month column
X_pay['day_in_m'] = X_pay.index.days_in_month 

## The pay days are every two weeks on the 15th and last day of month -- UNDERSTAND THIS
X_pay['pay_day'] = (X_pay.index.day == 16) | (X_pay.index.day == 1) | (X_pay.index.day == 14) | (X_pay.index.day == X_pay['day_in_m'] - 1) | (X_pay.index.day == 15) | (X_pay.index.day == X_pay['day_in_m'])

## Remove the days in month column as only needed for payday column
X_pay.drop(columns='day_in_m', inplace=True, axis=1)
```

#### Create a national holiday feature

```python
## Create a holiday dataframe
hol = holiday.loc['2017'].loc[holiday.loc['2017'].locale.isin(['National', 'Regional'])]

## Create a holiday column of 1s
hol['hol'] = 1      

## If holiday transferred, change to 0
hol.loc[(hol.type == 'Holiday') & (hol.transferred == True), 'hol'] = 0 

## Drop the other columns in holiday. Create dummies for other columns
hol = pd.get_dummies(hol.drop(columns=['locale','locale_name','description','transferred']), columns=['type'])

## Merge holiday dataframe with full dataframe
X_hol = X_pay.copy()
X_hol = pd.concat([X_hol, hol], axis=1).fillna(0)
X_hol.loc[X_hol.index.dayofweek.isin([5,6]), 'hol'] = 1
```

#### Create an oil feature 

```python
X_oil = X_hol.copy()
X_oil['oil'] = oils.rolling(7).mean()
```

### - Build the model -

#### Split dataset into features and labels
Note that the model is only trained on the 2017 data as most recent. 
Yearly fluctuations aren't accounted for.

```python
X_for_subm = X_oil.loc['2017':'2017-08-15'].dropna()
X_test = X_oil.loc['2017-08-16':'2017-08'].dropna()

myPrint(X_for_subm, printmessage="Final train data", hold=False)
myPrint(y_, printmessage="Full feature dataset", hold=False)
```

#### Fit the model

```python
model = LinearRegression()
model.fit(X_for_subm, y_) 
```

#### Create predictions for train data set

```python
y_pred = pd.DataFrame(model.predict(X_for_subm), columns=y_.columns, index=X_for_subm.index)
y_pred_ = y_pred.stack(['store_nbr', 'family'])
y_pred_.loc[y_pred_['sales'] < 0] = 0
ny_pred = y_pred_.unstack(['store_nbr', 'family'])
```

#### Plotting a few examples - data vs prediction

```python
plt.figure(figsize=(10,6));
plt.plot(y_.loc(axis=1)['sales',1, 'PRODUCE'].loc['2017'].values)
plt.plot(y_pred.loc(axis=1)['sales',1, 'PRODUCE'].loc['2017'].values)
plt.show()

plt.figure(figsize=(10,6));
plt.plot(y_.loc(axis=1)['sales',13, 'BOOKS'].loc['2017'].values)
plt.plot(y_pred.loc(axis=1)['sales',13, 'BOOKS'].loc['2017'].values)
plt.show()

plt.figure(figsize=(10,6));
plt.plot(y_.loc(axis=1)['sales',30, 'BEVERAGES'].loc['2017'].values)
plt.plot(y_pred.loc(axis=1)['sales',30, 'BEVERAGES'].loc['2017'].values)
plt.show()
```

#### Calculate Mean Absolute Error of model

```python
mae = mean_absolute_error(ny_pred.loc(axis=1)['sales',1, 'PRODUCE'].loc['2017'].values,
                          y_.loc(axis=1)['sales',1, 'PRODUCE'].loc['2017'].values)
msle = mean_squared_log_error(ny_pred.loc(axis=1)['sales',1, 'PRODUCE'].loc['2017'].values,
                              y_.loc(axis=1)['sales',1, 'PRODUCE'].loc['2017'].values)
print('mae = {:.3f}\nrmsle = {:.3f}'.format(mae, np.sqrt(msle)))
```

### - Create submission file for Kaggle competition using just multivariate Linear Regression -

```python
y_submit = pd.DataFrame(model.predict(X_test), columns=y_.columns, index=X_test.index)
y_submit_ = y_submit.stack(['store_nbr', 'family'])
y_submit_.loc[y_submit_['sales'] < 0] = 0

df_test_ = df_test.set_index(['store_nbr', 'family', 'date']).sort_index()
y_submit_ = y_submit_.join(df_test_.id).reindex(columns=['id', 'sales'])
#y_submit_.to_csv('submission_SLR_3.csv', index=False)
```

### - XGBoost training - 
Regression good for trends and extrapolation, but the model is linear. It can't pick up nonlinear features/interactions
or capture the error aspect in a MVLR. Therefore, we boost the model by fitting the residual noise.
The other option is stacking where we train a model using the predictions as features.

#### Create a residual. Need to move from stacked to unstacked form.

```python
y_gbm_ = y_.unstack().to_frame() - y_pred.unstack().to_frame()
myPrint(y_gbm_, printmessage="ygbm", hold=False)
```

```python
## Rather than for the MVLR where we set all the features for the train and test set at once, here we need to set them
## seperately because we use the date as an index with the store and family as features, which means the date index is now
## not unique. This means we can't split the data by date.
```

```python
store_sales=df_train.set_index(['date']).loc['2017']
X_xgb = store_sales.drop('sales', axis=1)

nX_xgb= X_xgb.set_index(['store_nbr', 'family', X_xgb.index]).sort_index()
myPrint(nX_xgb,hold=False)


le = LabelEncoder()
X_xgb['family'] = le.fit_transform(X_xgb['family'])

X_xgb['store_nbr'] = le.fit_transform(X_xgb['store_nbr'])

X_xgb["day"] = X_xgb.index.day  # values are day of the month
X_xgb.drop(columns=['id'], inplace=True)

X_xgb = X_xgb.loc['2017']
myPrint(X_xgb, printmessage="X_xgb", hold=False)
```

```python
X_xgb_test  = df_test.set_index(['store_nbr', 'family', 'date']).sort_index()
X_xgb_test = X_xgb_test.reset_index(['family', 'store_nbr'])

check_X_xgb_test = df_test.set_index(['date'])

X_xgb_test=check_X_xgb_test
X_xgb_test['family'] = le.fit_transform(X_xgb_test['family'])

X_xgb_test['store_nbr'] = le.fit_transform(X_xgb_test['store_nbr'])

## values are day of the month
X_xgb_test["day"] = X_xgb_test.index.day 
X_xgb_test.drop(columns=['id'], inplace=True)

X_xgb_test = X_xgb_test.loc['2017']
myPrint(X_xgb_test, printmessage="X_xgb_test", hold=False)
```

### - Notes on sorting label issues when moving from MVLR to XGBoost or other models -

While the linear regression algorithm is capable of multi-output regression, 
the XGBoost algorithm is not. To predict multiple series at once with XGBoost, 
we'll instead convert these series from wide format, with one time series per column, 
to long format, with series indexed by categories along rows.

So that XGBoost can learn to distinguish our two time series, we'll turn the row labels 
for 'Industries' into a categorical feature with a label encoding. We'll also create a feature 
for annual seasonality by pulling the month numbers out of the time index.

Now we'll convert the trend predictions made earlier to long format and then subtract them 
from the original series. That will give us detrended (residual) series that XGBoost can learn.

#### Train the model

```python
model_2 = XGBRegressor()
```

```python
model_2.fit(X_xgb, y_gbm_)
gbm_pred_2 = pd.DataFrame(model_2.predict(X_xgb), index=X_xgb.index, columns=['sales'])
myPrint(gbm_pred_2, printmessage="predicted 2", hold=False)
```

#### Add the XGboost predicted values to the MVLR predictions

```python
y_boost = pd.DataFrame(y_pred.unstack().to_frame().values + gbm_pred_2.values, index=nX_xgb.index, columns=['sales'])
y_boost.loc[y_boost.sales < 0] = 0
myPrint(y_boost, printmessage="Full prediction", hold=False)
```

#### Plotting a few examples - data vs prediction

```python
myPrint(y_boost, hold=False)
plt.figure(figsize=(20,6));
plt.plot(y_.loc(axis=1)['sales',1, 'PRODUCE'].loc['2017'].values)
plt.plot(y_boost.loc(axis=0)[1, 'PRODUCE'].loc['2017'].values)
plt.show()

plt.figure(figsize=(20,6));
plt.plot(y_.loc(axis=1)['sales',13, 'BOOKS'].loc['2017'].values)
plt.plot(y_boost.loc(axis=0)[13, 'BOOKS'].loc['2017'].values)
plt.show()

plt.figure(figsize=(20,6));
plt.plot(y_.loc(axis=1)['sales',30, 'BEVERAGES'].loc['2017'].values)
plt.plot(y_boost.loc(axis=0)[30, 'BEVERAGES'].loc['2017'].values)
plt.show()
```

#### Mean absolute error and mean squared log error for one example

```python
mae = mean_absolute_error(y_boost.loc(axis=0)[1, 'PRODUCE'].loc['2017'].values,
                          y_.loc(axis=1)['sales',1, 'PRODUCE'].loc['2017'].values)
msle = mean_squared_log_error(y_boost.loc(axis=0)[1, 'PRODUCE'].loc['2017'].values,
                              y_.loc(axis=1)['sales',1, 'PRODUCE'].loc['2017'].values)
print('mae = {:.3f}\nrmsle = {:.3f}'.format(mae, np.sqrt(msle)))
```

#### Predictions on test set

```python
gbm_pred_2 = pd.DataFrame(model_2.predict(X_xgb_test), index=y_submit_.index, columns=['sales'])
y_boost_2 = pd.DataFrame(y_submit.unstack().to_frame().values + gbm_pred_2.values, index=gbm_pred_2.index, columns=['sales'])
myPrint(y_boost_2, printmessage="Boosted submission predictions", hold=False)
```
